Preprocessing OpenAlex

```python
import requests
import requests
import json
import pandas as pd
import random
import time
import os
from collections import defaultdict

random.seed(42)

# output path for the final CSV
OUTPUT_PATH = r"C:\Users\hp\Desktop\Research-System\training\datasets\dataset_openalex.csv"

# how many papers we want per subcategory
CAP_PER_SUB = 20000

# polite delay between API requests in seconds
# OpenAlex allows 10 requests/second without an API key
REQUEST_DELAY = 0.15

REQUEST_DELAY = 0.15
MAX_RETRIES = 5
```

```python
def reconstruct_abstract(inverted_index: dict) -> str:
    # inverted index maps each word to a list of positions it appears at
    # we need to reverse this into a position -> word mapping
    # then sort by position and join into a string

    if not inverted_index:
        return ""

    # build a list of (position, word) pairs
    # for each word, it may appear at multiple positions so we loop through all of them
    position_word_pairs = []
    for word, positions in inverted_index.items():
        for pos in positions:
            position_word_pairs.append((pos, word))

    # sort by position so words appear in the correct order
    position_word_pairs.sort(key=lambda x: x[0])

    # extract just the words in order and join with spaces
    return " ".join(word for pos, word in position_word_pairs)
```

```python
# maps each of our subcategory labels to one or more OpenAlex subfield IDs
# most subcategories map to a single subfield ID
# Chemical Engineering pools 3 subfields together to form one subcategory
# the main_label is always the parent category name

OPENALEX_CATEGORY_MAP = {

    # Chemistry
    "Organic Chemistry": {
        "main_label": "Chemistry",
        "sub_label": "Organic Chemistry",
        "subfield_ids": ["1605"]
    },
    "Inorganic Chemistry": {
        "main_label": "Chemistry",
        "sub_label": "Inorganic Chemistry",
        "subfield_ids": ["1604"]
    },
    "Analytical Chemistry": {
        "main_label": "Chemistry",
        "sub_label": "Analytical Chemistry",
        "subfield_ids": ["1602"]
    },
    "Physical and Theoretical Chemistry": {
        "main_label": "Chemistry",
        "sub_label": "Physical and Theoretical Chemistry",
        "subfield_ids": ["1606"]
    },
    "Spectroscopy": {
        "main_label": "Chemistry",
        "sub_label": "Spectroscopy",
        "subfield_ids": ["1607"]
    },
    "Electrochemistry": {
        "main_label": "Chemistry",
        "sub_label": "Electrochemistry",
        "subfield_ids": ["1603"]
    },

    # Engineering
    "Electrical & Electronic Engineering": {
        "main_label": "Engineering",
        "sub_label": "Electrical & Electronic Engineering",
        "subfield_ids": ["2208"]
    },
    "Civil & Structural Engineering": {
        "main_label": "Engineering",
        "sub_label": "Civil & Structural Engineering",
        "subfield_ids": ["2205"]
    },
    "Mechanical Engineering": {
        "main_label": "Engineering",
        "sub_label": "Mechanical Engineering",
        "subfield_ids": ["2210"]
    },
    "Chemical Engineering": {
        "main_label": "Engineering",
        "sub_label": "Chemical Engineering",
        # pooling 3 subfields: Fluid Flow, Catalysis, Process Chemistry
        "subfield_ids": ["1507", "1503", "1508"]
    },
    "Industrial & Manufacturing Engineering": {
        "main_label": "Engineering",
        "sub_label": "Industrial & Manufacturing Engineering",
        "subfield_ids": ["2209"]
    },

    # Economics & Business
    "Econometrics": {
        "main_label": "Economics & Business",
        "sub_label": "Econometrics",
        "subfield_ids": ["2002"]
    },
    "Finance": {
        "main_label": "Economics & Business",
        "sub_label": "Finance",
        "subfield_ids": ["2003"]
    },
    "Marketing": {
        "main_label": "Economics & Business",
        "sub_label": "Marketing",
        "subfield_ids": ["1406"]
    },
    "Management": {
        "main_label": "Economics & Business",
        "sub_label": "Management",
        "subfield_ids": ["1408"]
    },
    "Accounting": {
        "main_label": "Economics & Business",
        "sub_label": "Accounting",
        "subfield_ids": ["1402"]
    },
    "Management Information Systems": {
        "main_label": "Economics & Business",
        "sub_label": "Management Information Systems",
        "subfield_ids": ["1404"]
    },
}
```

```python
def fetch_papers_for_subfield(subfield_id):
    # this function fetches papers from the OpenAlex API for one subfield ID
    # it uses cursor-based pagination to walk through pages of 200 results each
    # it stops early once it has collected CAP_PER_SUB * 2 raw papers (40K)
    # so we have a large pool to sample from after pooling multiple subfields
    # returns a list of dicts, each with keys: title, abstract, source

    papers = []

    # cursor="*" is how OpenAlex starts cursor-based pagination
    # after the first page, we replace this with the next_cursor value from the response
    cursor = "*"

    # early stop threshold: collect at most double the cap before stopping
    # this gives us enough raw papers to sample from while not fetching millions
    # from large subfields like Organic Chemistry (3.5M works in OpenAlex)
    early_stop = CAP_PER_SUB * 2

    # the select parameter tells OpenAlex to only return the fields we actually need
    # omitting unused fields makes each response smaller and faster to parse
    select_fields = "title,abstract_inverted_index,primary_topic"

    print(f"  Fetching subfield {subfield_id}...")

    # keep fetching pages until we hit the early stop limit or run out of results
    while len(papers) < early_stop:

        # build the full API request URL for this page
        # filter: only papers whose primary topic belongs to this subfield
        # has_abstract=true: skip papers with no abstract (we can't use them for training)
        # per_page=200: maximum allowed by OpenAlex — reduces number of round trips
        # cursor: current pagination position
        # select: only return the fields we need
        url = (
            f"https://api.openalex.org/works"
            f"?filter=primary_topic.subfield.id:{subfield_id},has_abstract:true"
            f"&per_page=200"
            f"&cursor={cursor}"
            f"&select={select_fields}"
        )

        # RETRY LOOP
        # OpenAlex occasionally drops connections mid-collection (RemoteDisconnected)
        # instead of crashing the entire run, we retry up to MAX_RETRIES times
        # each retry waits longer than the previous one (exponential backoff)
        # this gives the server time to recover before we try again
        response = None
        for attempt in range(MAX_RETRIES):
            try:
                # timeout=30 means if we get no response within 30 seconds, raise an error
                # without a timeout, a stalled connection would hang forever
                response = requests.get(
                    url,
                    headers={"User-Agent": "ResearchPaperSystem/1.0"},
                    timeout=30
                )

                # if we got a successful HTTP 200 response, break out of the retry loop
                if response.status_code == 200:
                    break

                # if we got a non-200 status (e.g. 429 rate limit, 500 server error)
                # log it and fall through to the backoff sleep below
                print(f"  HTTP {response.status_code} on attempt {attempt + 1} of {MAX_RETRIES}, retrying...")

            except (requests.exceptions.ConnectionError, requests.exceptions.Timeout) as e:
                # ConnectionError covers RemoteDisconnected and similar network failures
                # Timeout covers cases where the server is too slow to respond
                print(f"  Connection error on attempt {attempt + 1} of {MAX_RETRIES}: {e}")

            # exponential backoff: wait 2s after attempt 1, 4s after attempt 2, etc.
            # this prevents hammering the server immediately after it drops a connection
            wait_seconds = 2 ** (attempt + 1)
            print(f"  Waiting {wait_seconds}s before retry...")
            time.sleep(wait_seconds)

        else:
            # this else block runs if the for loop completed without hitting a break
            # meaning all MAX_RETRIES attempts failed — we stop fetching this subfield
            print(f"  All {MAX_RETRIES} retry attempts failed. Stopping this subfield early.")
            break

        # parse the JSON response body into a Python dict
        data = response.json()

        # results is the list of paper objects on this page
        results = data.get("results", [])

        # if results is empty, we've gone past the last page — stop paginating
        if not results:
            break

        # process each paper on this page
        for work in results:
            title = work.get("title")
            abstract_index = work.get("abstract_inverted_index")

            # skip papers missing a title or abstract
            # both are required for the SciBERT classifier input
            if not title or not abstract_index:
                continue

            # OpenAlex stores abstracts as an inverted index:
            # we reconstruct the original sentence by reversing the mapping:
            # build a list of (position, word) pairs, sort by position, join words
            # reconstruct_abstract() is defined in Block 2
            abstract = reconstruct_abstract(abstract_index)

            # skip papers where the reconstructed abstract is too short
            # very short abstracts are usually metadata errors or non-English text
            if len(abstract) < 100:
                continue

            # skip papers with a very short title — same reasoning as above
            if len(title.strip()) < 5:
                continue

            papers.append({
                "title": title.strip(),
                "abstract": abstract,
                "source": "openalex"
                # main_label and sub_label are attached in Block 5 after this function returns
                # we don't attach them here because this function only knows the subfield ID,
                # not the human-readable label — the category map in Block 5 has that mapping
            })

        print(f"  Collected {len(papers)} papers so far...")

        # get the cursor for the next page from the response metadata
        # when next_cursor is None, OpenAlex has no more pages for this subfield
        next_cursor = data.get("meta", {}).get("next_cursor")
        if not next_cursor:
            break

        # update cursor for the next iteration of the while loop
        cursor = next_cursor

        # polite delay between requests to avoid hitting the rate limit
        # OpenAlex allows 10 requests/second without an API key
        # 0.15s delay keeps us well under that limit
        time.sleep(REQUEST_DELAY)

    # report why we stopped
    if len(papers) >= early_stop:
        print(f"  Reached early stop limit ({early_stop}), moving on.")

    print(f"  Done. Total collected: {len(papers)}")
    return papers
```

```python
# MAIN COLLECTION LOOP WITH CHECKPOINTING
# How checkpointing works:
# After each subcategory finishes collecting, we immediately save its papers
# to a individual JSON file on disk (one file per subcategory).
# If the run crashes or gets interrupted, those files survive in memory.
# When we restart and re-run this block, we check at the top of the loop
# whether a checkpoint file already exists for that subcategory.
# If it does, we load it from disk and skip the API collection entirely.
# This means a crash only loses the subcategory currently being fetched —
# everything before it is safe on disk and gets reloaded automatically.

# Checkpoint files are saved to a dedicated folder so they don't clutter
# the main dataset directory. They are temporary — once Block 6 saves the
# final CSV, the checkpoint folder is no longer needed.

# folder where per-subcategory checkpoint files will be saved
# one JSON file per subcategory, named after the subcategory label
CHECKPOINT_DIR = r"C:\Users\hp\Desktop\Research-System\training\datasets\openalex_checkpoints"

# create the checkpoint folder if it doesn't already exist
os.makedirs(CHECKPOINT_DIR, exist_ok=True)

# papers_by_sub maps subcategory label → list of paper dicts
# using defaultdict so we never need to check if a key exists before appending
papers_by_sub = defaultdict(list)

for sub_name, entry in OPENALEX_CATEGORY_MAP.items():

    # build the path for this subcategory's checkpoint file
    # we replace spaces with underscores and & with 'and' so the filename is safe
    safe_name = sub_name.replace(" ", "_").replace("&", "and")
    checkpoint_path = os.path.join(CHECKPOINT_DIR, f"{safe_name}.json")

    # CHECK FOR EXISTING CHECKPOINT
    # if a checkpoint file exists for this subcategory, it means a previous run
    # already finished collecting it — load from disk and skip the API entirely
    if os.path.exists(checkpoint_path):
        print(f"\nResuming: {sub_name} — loading from checkpoint...")
        with open(checkpoint_path, "r", encoding="utf-8") as f:
            papers_by_sub[sub_name] = json.load(f)
        print(f"  Loaded {len(papers_by_sub[sub_name])} papers from checkpoint.")
        continue  # skip to the next subcategory

    print(f"\nCollecting: {sub_name}")

    # sub_papers accumulates raw papers from all subfield IDs for this subcategory
    # for most subcategories this list comes from one fetch call
    # for Chemical Engineering it comes from three fetch calls (IDs 1507, 1503, 1508)
    # that are pooled together before capping
    sub_papers = []

    for subfield_id in entry["subfield_ids"]:
        # fetch all available papers (up to early_stop = 40K) for this subfield
        fetched = fetch_papers_for_subfield(subfield_id)

        # attach the human-readable labels from the category map
        # we do this here rather than inside the fetcher because the fetcher
        # only knows the subfield ID number, not what label we've assigned to it
        for paper in fetched:
            paper["main_label"] = entry["main_label"]
            paper["sub_label"] = entry["sub_label"]

        sub_papers.extend(fetched)
        print(f"  Pooled total so far: {len(sub_papers)}")

    # after pooling all subfield IDs for this subcategory, cap at CAP_PER_SUB
    # random.sample picks exactly CAP_PER_SUB papers without replacement
    # this ensures no subcategory dominates the final dataset
    if len(sub_papers) > CAP_PER_SUB:
        sub_papers = random.sample(sub_papers, CAP_PER_SUB)
        print(f"  Capped to {CAP_PER_SUB}")
    else:
        # if we got fewer than the cap, warn so we know the final dataset
        # will be slightly unbalanced for this subcategory
        print(f"  WARNING: {sub_name} only has {len(sub_papers)} papers — using all")

    papers_by_sub[sub_name] = sub_papers
    print(f"  Final count for {sub_name}: {len(papers_by_sub[sub_name])}")

    # SAVE CHECKPOINT
    # write this subcategory's papers to disk immediately after capping
    # if the run crashes on the next subcategory, this one is already saved
    # and will be loaded from disk on the next restart instead of re-fetched
    with open(checkpoint_path, "w", encoding="utf-8") as f:
        json.dump(sub_papers, f, ensure_ascii=False)
    print(f"  Checkpoint saved to {checkpoint_path}")

print("\nAll subcategories collected.")
print(f"Total subcategories in papers_by_sub: {len(papers_by_sub)}")
for sub_name, papers in papers_by_sub.items():
    print(f"  {sub_name}: {len(papers):,} papers")
```

```python
# Combine all subcategory paper lists into one flat list
all_papers = []
for sub_name, papers in papers_by_sub.items():
    all_papers.extend(papers)

# Convert to DataFrame
df = pd.DataFrame(all_papers)

# Make sure source column is set (if not already in the dicts)
df["source"] = "openalex"

# Reorder columns to match arXiv/PubMed CSV format
df = df[["title", "abstract", "main_label", "sub_label", "source"]]

# Deduplicate by title + abstract (same paper can appear in multiple subfields)
before = len(df)
df = df.drop_duplicates(subset=["title", "abstract"]).reset_index(drop=True)
print(f"Duplicates removed: {before - len(df):,}")

# Quick null check
assert df["title"].isnull().sum() == 0, "Found null titles!"
assert df["abstract"].isnull().sum() == 0, "Found null abstracts!"

# Shuffle so subcategories are not grouped together
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Save
df.to_csv(OUTPUT_PATH, index=False)
print(f"Saved to {OUTPUT_PATH}")
print(f"Total papers: {len(df):,}")

# Distribution check
print("\nPapers per subcategory:")
print(df["sub_label"].value_counts().to_string())
```
